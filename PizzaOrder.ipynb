```
from datetime import date
import datetime
import random

storeIdReference = 353535
pizzaIdReference = 252525
orderIdReference = 100000
promotionalCodeId = 'GETOFF2'

stores = []
pizzas = []
selectedPizzas = []
promotionalCodes = {}
pizzaOrders = []

def main(): 
  AddStores()
  AddPizza()
  AddPromotionalCode()
  CreateDummyPizzaOrders()
  Order_Pizza()


def AddStores(): 
  global storeIdReference, stores
  address1 = Address('Street1', 5, '21227')
  store1 = Store(storeIdReference, address1, '+410-21565854', 500) 
  stores.append(store1)
  storeIdReference = storeIdReference + 1;

  employee1 = Employee('Tim','Paine')
  store1.AddEmployee(employee1);

  address2 = Address('Street2', 10, '21228')
  store2 = Store(storeIdReference, address2, '+410-85426321', 750) 
  stores.append(store2)
  storeIdReference = storeIdReference + 1;

  employee2 = Employee('Jennifer2','Anniston')
  store2.AddEmployee(employee2);

  address3 = Address('Street3', 4, '21250')
  store3 = Store(storeIdReference, address3, '+410-96542157', 450) 
  stores.append(store3)
  storeIdReference = storeIdReference + 1;

  employee3 = Employee('Jennifer3','Anniston')
  store3.AddEmployee(employee3);

  address4 = Address('Street4', 9, '21210')
  store4 = Store(storeIdReference, address4, '+410-96584212', 600) 
  stores.append(store4)
  storeIdReference = storeIdReference + 1;

  employee4 = Employee('Jennifer4','Anniston')
  store4.AddEmployee(employee4);

  address5 = Address('Street5', 1, '21220')
  store5 = Store(storeIdReference, address5, '+410-96584231', 300) 
  stores.append(store5)
  storeIdReference = storeIdReference + 1;

  employee5 = Employee('Jennifer5','Anniston')
  store5.AddEmployee(employee5)

def AddPizza():
  global pizzaIdReference, pizzas
  pizza1 = Pizza(pizzaIdReference, 'Margherita', '$10', 'Normal', 'THIN');
  pizzas.append(pizza1) 
  pizzaIdReference = pizzaIdReference + 1

  pizza2 = Pizza(pizzaIdReference, 'Farm House', '$12', 'Normal', 'THIN');
  pizzas.append(pizza2)
  pizzaIdReference = pizzaIdReference + 1

  pizza3 = Pizza(pizzaIdReference, 'Mexican Green Wave', '$15', 'Normal', 'THIN');
  pizzas.append(pizza3) 
  pizzaIdReference = pizzaIdReference + 1

  pizza4 = Pizza(pizzaIdReference, 'Veg Extravaganza', '$20', 'Normal', 'THIN');
  pizzas.append(pizza4)

def AddPromotionalCode():  
  global promotionalCodes
  promotionalCodes={'GETOFF2': 2, 'GETOFF5' : 5}

def CreateDummyPizzaOrders():
  global orderIdReference
  global promotionalCodes
  global stores
  global pizzas

  customers = []

  customer1 = Customer('Joey', 'Tribbiani', '+410-95896584', 21227, 1000);
  customers.append(customer1)

  customer2 = Customer('Ross', 'Geller', '+410-96542133', 21210, 650);
  customers.append(customer2)

  customer3 = Customer('Chandler', 'Bing', '+410-85463322', 21220, 500);
  customers.append(customer3)

  customer4 = Customer('Pheobe', 'Buffay', '+410-22114455', 21227, 750);
  customers.append(customer4)

  customer5 = Customer('Monica', 'Geller', '+410-22336655', 21228, 250);
  customers.append(customer5)

  customer6 = Customer('Rachel', 'Green', '+410-12345698', 21250, 450);
  customers.append(customer6)

  orderDates = []
    
  year, month, day = map(int, '2018-02-15'.split('-'))
  date1 = datetime.datetime(year, month, day)
  orderDates.append(date1)

  year, month, day = map(int, '2019-05-26'.split('-'))
  date2 = datetime.datetime(year, month, day)
  orderDates.append(date2)

  year, month, day = map(int, '2019-02-05'.split('-'))
  date3 = datetime.datetime(year, month, day)
  orderDates.append(date3)

  year, month, day = map(int, '2018-10-25'.split('-'))
  date4 = datetime.datetime(year, month, day)
  orderDates.append(date4)

  year, month, day = map(int, '2020-02-15'.split('-'))
  date5 = datetime.datetime(year, month, day)
  orderDates.append(date5)

  orderStatus=[]

  orderStatus.append('ORDER_CANCELED')
  orderStatus.append('ORDER_COMPLETE')
    
  orderIndex = 0
  while orderIndex < len(orderDates):
    store = stores[random.randint(0, len(stores) - 1)]

    customer = customers[random.randint(0, len(customers) - 1)]

    pizzaOrder = PizzaOrder(orderIdReference ,store, customer)
    orderIdReference = orderIdReference + 1

    count = random.randint(1, len(pizzas))
    
    index = 0
    while index < count:
      pizzaOrder.AddPizza(pizzas[index])
      pizzaOrder.UpdateCartValue((pizzas[index]).price[1:], 'add')
      index = index + 1

      doApplyPromotionalCode = False
      val = random.randint(0,1)
      if val == 1:
        doApplyPromotionalCode = True

      if doApplyPromotionalCode == True:
        promoCodeIndex = random.randint(0,len(promotionalCodes))
        currentIndex = 0
        code = str()
        for temp in promotionalCodes:
          code = temp
          if(currentIndex == promoCodeIndex):
            break
          currentIndex = currentIndex + 1
        pizzaOrder.ApplyPromotionalCode(code)
        pizzaOrder.UpdateCartValue(promotionalCodes[code], 'remove')
    pizzaOrder.orderDate = orderDates[orderIndex]
    pizzaOrder.orderStatus=orderStatus[random.randint(0,1)]

    pizzaOrders.append(pizzaOrder)
         #DisplayOrderDetails(pizzaOrder)
    orderIndex = orderIndex + 1     

def Order_Pizza():
    global pizzaOrders
    global orderIdReference
    global stores
    print("Stores:")
    print(' ')
    for store in stores:
        print('Store Number:' + str(store.storeId))
        print('Store Address:' + store.storeAddress.toString())
        print('Store Phone Number:' + store.phoneNumber)
        print('Store Monthly Sales:' + str(store.monthlySales))
        print(' ')

    global pizzas
    print("Available Pizzas:")
    print(' ')
    for pizza in pizzas:
        print('Pizza Id:' + str(pizza.id))
        print('Pizza Name' + pizza.name)
        print('Pizza topping' + pizza.topping)
        print('Pizza price:' + pizza.price)
        print('Pizza Crust Type:' + pizza.crust_Type)
        print(' ')

    global promotionalCodes
    print("Available promotional codes:")
    print(' ')
    for codes in promotionalCodes:
        print(codes +": Get $"+ str(promotionalCodes[codes])+ " off on total amount")
    print(' ')

    while True:
        print('Please select the store. Enter store Id')
        selectedStoreId = int(input())

        while (not isSelectedStoreIdExists(selectedStoreId)):
            print('Please select the store from the ones listed above :)')
            selectedStoreId = int(input())

        print('Please enter the Customer details')
        print('First Name:')
        firstName = str(input())
        print('Last Name:')
        lastName = str(input())
        print('Phone Number:')
        phoneNumber = str(input())
        print('Zip Code:')
        zipCode = int(input())
        print('Frequent Mileage Number:')
        frequentMileageNumber = int(input())

        customer = Customer(firstName, lastName, phoneNumber, zipCode, frequentMileageNumber)

        while True:
            print('Please select the Pizza .Enter Pizza Id or to checkout, press 1. To remove any selected pizza, please press 2')
            pizzaID=int(input())
          
            if pizzaID == 1:
                break
            
            if pizzaID==2:
                pizzaIdToRemove=input("Enter the PizzaID to remove")
                
                while(not isSelectedPizzaIdExists(pizzaIdToRemove)):
                    pizzaIdToRemove=input("Invalid Input. Please select the Pizza .Enter PizzaId. Press 1 to cancel operation")
                    
                    if pizzaIdToRemove == 1:
                        break

                for aPizza in pizzas:
                    if(aPizza.id == pizzaIdToRemove):
                        selectedPizzas.remove(aPizza)
            
            else:
                while(not isSelectedPizzaIdExists(pizzaID)):
                    print("Invalid Input. Please select the Pizza .Enter PizzaId. Press 1 to cancel operation")
                    pizzaID=input()
                
                    if pizzaID==1:
                        break

                for aPizza in pizzas:
                    if(aPizza.id == pizzaID):
                        selectedPizzas.append(aPizza)

        print("Apply Promotional Code: Please enter the promotional code(is any) or else press E to continue to checkout")
        
        promotionalCode = str(input())
            
        applyPromotionalCode = False

        while True:
            if promotionalCode == "E":
                break
                
            #if promotionalCodes.has_key(promotionalCode):
            if promotionalCode in promotionalCodes: 
                applyPromotionalCode = True
                break
                
            print("Invalid promotional code entered. Please enter the correct promotional code or press E to continue to checkou")         
            promotionalCode = str(input())

        for store in stores:
            if (store.storeId == selectedStoreId):
                pizzaOrder = PizzaOrder(orderIdReference ,store, customer)
                orderIdReference = orderIdReference + 1

                for orderedPizza in selectedPizzas:
                    pizzaOrder.AddPizza(orderedPizza)
                    pizzaOrder.UpdateCartValue(orderedPizza.price[1:], 'add')

                if applyPromotionalCode == True:
                    pizzaOrder.ApplyPromotionalCode(promotionalCode)
                    pizzaOrder.UpdateCartValue(promotionalCodes[promotionalCode], 'remove')

                pizzaOrders.append(pizzaOrder)
                DisplayOrderDetails(pizzaOrder)          
                break

        continueWithOrder=1
        continueWithOptionSelection=1

        while continueWithOptionSelection == 1:
            
            option=str(input("Press C to continue for next order, Q for querying orders or E to Exit"))
            
            if option=='C':
                continueWithOptionSelection = 0
                continueWithOrder = 1
                break
            elif option=='E':
                continueWithOptionSelection = 0
                continueWithOrder = 0
                break
            elif option=='Q':
                QueryOrders()
                continueWithOptionSelection = 1
                continueWithOrder = 0
            else:
                continueWithOptionSelection = 1
                print("Invalid Option")
                #break 
        
        if continueWithOrder==0:
          break

def QueryOrders():
    print('''
        Sort PizzaOrder by order date and total order amount: Press 1 
        Search PizzaOrder by customer: Press 2,
        Search PizzaOrder by order date: Press 3,
        Pull out a list of PizzaOrder prior to a certain date in sorted order by date: Press 4,
        Pull out a list of PizzaOrder after a certain date in sorted order by date: Press 5,
        Find one of your stores in the same zip code: Press 6''')
    queryInput=int(input("Enter Query number"))
    print("You entered Query Number: ", queryInput)
    if queryInput==1:
      #print("jkkkjh")
      newPizzaOrdersList1=[]
      newPizzaOrdersList1=sorted(pizzaOrders, key=lambda x:x.orderDate) 
      newPizzaOrdersList1=sorted(newPizzaOrdersList1, key=lambda x:x.totalAmount)
      #This feature is for the company to see all the orders in sorted order      
      print("Sorted Pizza Orders by Order Date and Total Amount is: ")
      for i in newPizzaOrdersList1:
        DisplayOrderDetails(i)

    elif queryInput==2:
      #Search PizzaOrder by customer
      firstName= str(input("To select the Customer enter the First Name"))
      
      newPizzaOrdersList2 = []

      for i in pizzaOrders:
        if firstName == i.customer.FirstName:
            newPizzaOrdersList2.append(i)

      for i in newPizzaOrdersList2:
        DisplayOrderDetails(i) 

    elif queryInput==3:
        #Search PizzaOrder by order date
        newPizzaOrdersList3=[]
        searchingDate=str(input('''Which date's order are you looking for? 
                          Please enter the date in YYYY-MM-DD format'''))
        year, month, day = map(int, searchingDate.split('-'))
        InputDate = datetime.date(year, month, day)

        for i in pizzaOrders:
          if(type(i.orderDate) is datetime.datetime):
                if i.orderDate < datetime.datetime(InputDate.year, InputDate.month, InputDate.day):
                    newPizzaOrdersList3.append(i)
          else:
                if datetime.datetime(i.orderDate.year, i.orderDate.month, i.orderDate.day) < datetime.datetime(InputDate.year, InputDate.month, InputDate.day):
                    newPizzaOrdersList3.append(i)
            #if InputDate == i.orderDate:
                #newPizzaOrdersList3.append(i)
      
        for i in newPizzaOrdersList3:
            DisplayOrderDetails(i)

    elif queryInput==4:

        #list of PizzaOrder prior to a certain date in sorted order by date
        newPizzaOrdersList4=[]
        #newPizzaOrdersList4=sorted(pizzaOrders, key=lambda x:x.orderDate)
        InputDate1=str(input('''Prior to which date do you want the Pizza Orders to be listed?
                        Please enter the date in YYYY-MM-DD format '''))
        year, month, day = map(int, InputDate1.split('-'))
        InputDate1 = datetime.datetime(year, month, day)
      
        for i in pizzaOrders:

            if(type(i.orderDate) is datetime.datetime):
                if i.orderDate < datetime.datetime(InputDate1.year, InputDate1.month, InputDate1.day):
                    newPizzaOrdersList4.append(i)
            else:
                if datetime.datetime(i.orderDate.year, i.orderDate.month, i.orderDate.day) < datetime.datetime(InputDate1.year, InputDate1.month, InputDate1.day):
                    newPizzaOrdersList4.append(i)
      
        for i in newPizzaOrdersList4:
            DisplayOrderDetails(i)

    elif queryInput==5:

        newPizzaOrdersList5=[]
        #Pull out a list of PizzaOrder after a certain date in sorted order by date
        InputDate2=str(input('''From which date onwards do you want the Pizza Orders to be listed?
                        Please enter the date in YYYY-MM-DD format '''))
        year, month, day = map(int, InputDate2.split('-'))
        InputDate2 = datetime.datetime(year, month, day)
        
        for i in pizzaOrders:

            if(type(i.orderDate) is datetime.datetime):
                if i.orderDate > datetime.datetime(InputDate2.year, InputDate2.month, InputDate2.day):
                    newPizzaOrdersList5.append(i)
            else:
                if datetime.datetime(i.orderDate.year, i.orderDate.month, i.orderDate.day) > datetime.datetime(InputDate2.year, InputDate2.month, InputDate2.day):
                    newPizzaOrdersList5.append(i)

        for i in newPizzaOrdersList5:
            DisplayOrderDetails(i)

    elif queryInput==6:
      #Find one of your stores in the same zip code
      InputZipCode=str(input("Please enter the zip code where you are searching a Pizza Store"))
      for i in stores:
       if InputZipCode in i.storeAddress.zipCode:
        print('Store:')
        print(' ')
        print('Store Number: ', str(i.storeId))
        print('Store Address: ', i.storeAddress.toString())
        print('Store Phone Number: ', i.phoneNumber)
        print('Store Monthly Sales: ', str(i.monthlySales))
    else:
      print("Incorrect Query Number")
     
def DisplayOrderDetails(pizzaOrder):
    print(' ')

    print('Order Details:')

    print('Order Date: ', pizzaOrder.orderDate)
    print('Order Status', pizzaOrder.orderStatus)
    print(' ')
    print('Store:')
    print(' ')
    print('Store Number: ' + str(pizzaOrder.store.storeId))
    print('Store Address: ' + pizzaOrder.store.storeAddress.toString())
    print('Store Phone Number: ' + pizzaOrder.store.phoneNumber)
    print('Store Monthly Sales: ' + str(pizzaOrder.store.monthlySales))
    print(' ')

    print("Selected Pizzas:")
    print(' ')
    for pizza in pizzaOrder.pizzasOrdered:
        print('Pizza Id: ' + str(pizza.id))
        print('Pizza Name: ' + pizza.name)
        print('Pizza topping: ' + pizza.topping)
        print('Pizza price: ' + pizza.price)
        print('Pizza Crust Type: ' + pizza.crust_Type)
        print(' ')

    #if(promotionalCodes.has_key(pizzaOrder.promotionalCode)):
    if pizzaOrder.promotionalCode in promotionalCodes:
        print('Applied promotional code: ')
        print(pizzaOrder.promotionalCode +": $"+ str(promotionalCodes[pizzaOrder.promotionalCode])+ " off on total amount")
    print(' ')

    #Console.WriteLine("Customer:");
    #DisplayCustomerInfo(pizzaOrder.Customer);

    print("Total Cart value: $" + str(pizzaOrder.GetCartValue()))    

def isSelectedStoreIdExists(selectedStoreId):
    global stores
    
    for store in stores:
        if (store.storeId == selectedStoreId):
            return True

    return False

def isSelectedPizzaIdExists(pizzaId):
    global pizzas

    for piza in pizzas:
        if(piza.id == pizzaId):
            return True

    return False

#if __name__ == "__main__":
#    main()
class PizzaOrder:


    def __init__(self, orderIdReference, store, customer):
        self.orderId = orderIdReference
        self.pizzasOrdered = []
        self.store = store
        self.orderStatus = 'ORDER_CREATED'
        self.customer = customer
        self.orderDate = datetime.datetime.now()
        self.ordertime = datetime.time
        self.totalAmount = 0
        self.promotionalCode = str()

    def GetCartValue(self):
        return self.totalAmount

    def AddPizza(self, apizza):
        self.pizzasOrdered.append(apizza)

    def RemovePizza(self, aPizza):
        self.pizzasOrdered.remove(aPizza);

    def ApplyPromotionalCode(self, code):
        self.promotionalCode = code

    def GetOrderStatus(self):
        return self.orderStatus

    def UpdateCartValue(self, value, operation):
        if operation.lower() == 'add':
            self.totalAmount = self.totalAmount + int(value)
        else:
            self.totalAmount = self.totalAmount - int(value)

class Pizza:
  def __init__(self, id, name, price, topping, crust_Type):
    self.id = id
    self.name = name
    self.price= price
    self.topping = topping
    self.crust_Type = crust_Type    

  def AddTopping(self, topping):
    self.topping = topping

  def RemoveTopping(self):
    self.topping = str()

  def SpecifyPrice(self, price):
    self.price = price

  def SpecifyCrust(self, crust_Type):
    self.crust_Type = crust_Type

class Store:

    def __init__(self, storeId, storeAddress, phoneNumber, monthlySales):
        self.storeId = storeId
        self.storeAddress = storeAddress
        self.phoneNumber = phoneNumber
        self.monthlySales = monthlySales
        self.employee = []
    
    def AddEmployee(self, employee):
        self.employee.append(employee)

    def RemoveEmployee(self, employee):
        self.employee.remove(employee)

class Address:

    def __init__(self, streetName, number, zipCode):
        self.streetName = streetName
        self.number = number
        self.zipCode = zipCode

    def toString(self):
     return self.streetName + ', ' + str(self.number) + ', ' + self.zipCode

class Employee:

    def __init__(self, FirstName, LastName):
        self.FirstName = FirstName
        self.LastName = LastName

class Customer:

    def __init__(self, FirstName, LastName, PhoneNumber, ZipCode, FrequentMileageNumber):
        self.FirstName = FirstName
        self.LastName = LastName
        self.PhoneNumber = PhoneNumber
        self.ZipCode = ZipCode
        self.FrequentMileageNumber= FrequentMileageNumber

main()
```
