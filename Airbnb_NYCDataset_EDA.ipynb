```
boroname_count=df.groupby('boroname').agg('count').reset_index()
```

```

#Plot the count by borough
fig, ax1 = plt.subplots(1,1, figsize=(6,6)
                       )
sns.barplot(x='boroname', y='id', data=boroname_count, ax=ax1)

ax1.set_title('Number of Listings by Boroname', fontsize=15)
ax1.set_xlabel('Boroname', fontsize=12)
ax1.set_ylabel('Count', fontsize=12)
ax1.tick_params(axis='both', labelsize=10)
```
